### Question 1
* This code takes in a input of times in the format specified, then runs computations to see if any of the times overlap any of the other times
* using dateparser I can compare times easily its kinda a mess but i belive its a working mess

```python
import dateparser
list = []
n = int(input("Enter number of times: "))
for i in range(n):
    list.append(input("Enter time range in form HH:mmp-HH:mmp "))


split_times = [time.split("-") for time in list]
parsed_times = []
for i in range(len(split_times)):
    for j in range(len(split_times[i])):
        datetime = dateparser.parse(split_times[i][j], date_formats=['%I:%M%p'])
        parsed_times.append(datetime)
for i in range(0, len(parsed_times), 2):
    time1 = parsed_times[i]
    time2 = parsed_times[i + 1]
    for j in range(0, len(parsed_times), 2):
        target1 = parsed_times[j]
        target2 = parsed_times[j + 1]
        if (time1 <= target1 <= time2 or time1 <= target2 <= time2) and i != j:
            print(
                f'{time1:%I:%M%p}' + "-" + f'{time2:%I:%M%p}' + " overlaps with " + f'{target1:%I:%M%p}' + "-" + f'{target2:%I:%M%p}') 
```

### Question 2
* man was this a challenge in a half. had some issues with the key and the proper syntax for that but i did manage to get it working
* this code defines 2 methods main and weighted sum. main calls a sorted fucntion on list_data wich uses weighted sum as a key which uses the total to sort each country in the list. 

```python
list_data = [['Australia', 1, 2, 1, 4],
        ['Austria', 7, 7, 4, 18],
        ['Belarus', '-', 2, '-', 2],
        ['Belgium', 1, '-', 1, 2],
        ['Canada', 4, 8, 14, 26],
        ['Czech Republic', 1, '-', 1, 2],
        ['Estonia', '-', '-', 1, 1],
        ['Finland', 2, 2, 4, 8],
        ['France', 5, 7, 2, 14],
        ['Germany', 12, 10, 5, 27],
        ['Great Britain', 1, 1, '-', 2],
        ['Hungary', 1, '-', 2, 3],
        ['Italy', 2, 7, 8, 17],
        ['Japan', 3, 6, 9, 18],
        ['Latvia', '-', '-', 1, 1],
        ['Netherlands', 8, 5, 4, 17],
        ['New Zealand', 2, 1, '-', 3],
        ['Norway', 16, 8, 13, 37],
        ["People's Republic of China", 9, 4, 2, 15],
        ['Poland', '-', '-', 1, 1],
        ['ROC', 6, 12, 14, 32],
        ['Republic of Korea', 2, 5, 2, 9],
        ['Slovakia', 1, '-', 1, 2],
        ['Slovenia', 2, 3, 2, 7],
        ['Spain', '-', 1, '-', 1],
        ['Sweden', 8, 5, 5, 18],
        ['Switzerland', 7, 2, 5, 14],
        ['Ukraine', '-', 1, '-', 1],
        ['United States of America', 8, 10, 7, 25], ]
norm = [1, 1, 1]
fib = [3, 2, 1]
expo = [4, 2, 1]
topend = [6, 2, 1]


def weighted_sum(data: list, rank_vector: list):
    total = 0
    for i in range(len(data)):
        try:
            total += data[i] * rank_vector[i]
        except TypeError:
            pass

    return total


def main():
    norm_list = sorted(list_data, key=lambda row: weighted_sum(row[1:4], norm), reverse=True)
    fib_list = sorted(list_data, key=lambda row: weighted_sum(row[1:4], fib), reverse=True)
    expo_list = sorted(list_data, key=lambda row: weighted_sum(row[1:4], expo), reverse=True)
    topend_list = sorted(list_data, key=lambda row: weighted_sum(row[1:4], topend), reverse=True)
    
    print(f'{"Medal Order"}')
    for i in range(len(norm_list)):
        print(str(i) + ": " + str(norm_list[i]))
    
    print(f'{"Fibonacci Weighted: "}')
    for i in range(len(fib_list)):
        print(str(i) + ": " + str(fib_list[i]))
    
    print(f'{"Exponential Weighted: "}')
    for i in range(len(expo_list)):
        print(str(i) + ": " + str(expo_list[i]))
    
    print(f'{"Topend Sports: "}')
    for i in range(len(topend_list)):
        print(str(i) + ": " + str(topend_list[i]))


main()

```
